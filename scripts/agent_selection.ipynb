```python
import matplotlib.pyplot as plt

# Read the file
with open('/home/memedi/output2.txt', 'r') as file:
    data = file.read()

# Extract the values and flatten the list of lists
values = [int(num) for line in data.strip().split('\n') for num in line.strip('[]').split(', ')]

# Plot the histogram
plt.hist(values, bins=[0, 1, 2, 3], edgecolor='black', align='left')
plt.xticks([0, 1, 2])
plt.xlabel('Acton (i.e., Resource)')
plt.ylabel('Frequency')
plt.title('Frequency of selection')
plt.show()
```
